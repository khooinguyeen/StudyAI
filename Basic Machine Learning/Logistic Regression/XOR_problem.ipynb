```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
x = np.array([[0, 1], [1, 0], [0, 0], [1, 1]])
y = np.array([1, 1, 0, 0])

print(x, '\n', y)
```

```python
plt.scatter(x[:2, 0], x[:2, 1], c='m', edgecolors='none', s=30, label='output = 1')
plt.scatter(x[2:, 0], x[2:, 1], c='y', edgecolors='none', s=30, label='output = 0')
plt.legend(loc='center')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
model = LogisticRegression()
model.fit(x, y)
```

```python
wg = np.zeros( (3, 1) )
wg[0, 0] = model.intercept_
wg[1:, 0] = model.coef_
```

```python
t = 0.5
plt.scatter(x[:2, 0], x[:2, 1], c='m', edgecolors='none', s=30, label='output = 1')
plt.scatter(x[2:, 0], x[2:, 1], c='y', edgecolors='none', s=30, label='output = 0')
plt.legend(loc='center')
plt.xlabel('x1')
plt.ylabel('x2')
plt.plot((0, 1),(-(wg[0]+0*wg[1]+ np.log(1/t-1))/wg[2], \
-(wg[0] + 1*wg[1]+ np.log(1/t-1))/wg[2]), 'g')
plt.show()
```
