```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numOfPoint = 30
noise = np.random.normal(0, 1, numOfPoint).reshape(-1, 1)
print(noise)
```

```python
x = np.linspace(30, 100, numOfPoint).reshape(-1, 1)
print(x)
```

```python
N = x.shape[0]
print(N)
```

```python
y = 15 * x + 8 + 20 * noise
print(y)
```

```python
plt.scatter(x, y)
plt.show()
```
