```python
data = pd.read_csv('data_square.csv').values
N = data.shape[0]
x = data[:, 0].reshape(-1, 1)
y = data[:, 1].reshape(-1, 1)
plt.scatter(x, y)
plt.xlabel('m√©t vu√¥ng')
plt.ylabel('gi√°')
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)
plt.plot((x[0], x[-1]), (y_pred[0], y_pred[-1]), 'g')
plt.show()
np.savez('w3.npz', a=model.intercept_, b=model.coef_)

k = np.load('w3.npz')
model.intercept_ = k['a']
model.coef_ = k['b']
```
